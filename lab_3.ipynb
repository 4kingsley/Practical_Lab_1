```python
import hashlib

W = hashlib.sha3_256("Blockchain".encode()).hexdigest()
F = "Buterin"

for X in range(1000000, 9999999):
    text = W + F + str(X)
    h = hashlib.sha256(text.encode()).hexdigest()

    if h.startswith("cafe"):
        print("Found!")
        print("X =", X)
        print("Hash =", h)
        break
```
